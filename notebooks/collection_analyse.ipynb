With this notebook I'm trying to see how to manage collection with schedules (to store in a postgresql database)

```python
import json
```

```python
with open("data/collection.json", "r") as f:
    data = json.load(f)
data
```

```python
LocationCollection = data["samplingLocationCollections"][0]
location = LocationCollection.copy()
del location["collections"]
location
```

```python
collections = LocationCollection["collections"].copy()
collections
```

```python
cl1 = collections[0].copy()
cl1_ch = cl1["children"]
del cl1["children"]
```

```python
print(f"{'KEY':>20}\t{'VALUE_A':>20}\t{'VALUE_B':>20}\t{'IS_EQUAL':>10}")
for k, v in cl1.items():
    try:
        print(
            f"{k + ':':>20}\t{str(v)[:20]:>20}\t{str(cl1_ch[0][k])[:20]:>20}\t{v == cl1_ch[0][k]:>10}"
        )
    except Exception as e:
        print(f"key: {k}, value: {v}, error: {e}")
        break
```

```python
import pandas as pd

df = pd.DataFrame([cl1, *cl1_ch])
df
```

```python
for col in df.columns:
    print(f"{col}: {df[col].unique()}")
```
